```python
def plot_dihedral(data_298k, data_400k, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data_298k['Frame'], data_298k['Phi'], label='298K Phi', color='blue')
    plt.plot(data_298k['Frame'], data_298k['Psi'], label='298K Psi', color='skyblue')
    plt.plot(data_400k['Frame'], data_400k['Phi'], label='400K Phi', color='red')
    plt.plot(data_400k['Frame'], data_400k['Psi'], label='400K Psi', color='salmon')
    plt.title(f'ngulos diedros de {title}', size = 20)
    plt.xlabel('fentosegundos', size = 20)
    plt.ylabel('ngulo (grados)', size = 20)
    plt.legend()
    plt.grid(False)
    plt.show()

# Cargar los datos
ala_2_298k_data = pd.read_csv('tarea/4-analysis/ala-2_298k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])
ala_2_400k_data = pd.read_csv('tarea/4-analysis/ala-2_400k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])
gln_3_298k_data = pd.read_csv('tarea/4-analysis/gln-3_298k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])
gln_3_400k_data = pd.read_csv('tarea/4-analysis/gln-3_400k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])

# Graficar
plot_dihedral(ala_2_298k_data, ala_2_400k_data, 'ALA-2')
plot_dihedral(gln_3_298k_data, gln_3_400k_data, 'GLN-3')
```

Las grficas proporcionadas muestran los ngulos dihedros  (Phi) y  (Psi) para dos residuos diferentes (ALA-2 y GLN-3) de un tripptido AQA a dos temperaturas distintas (298K y 400K). Estos ngulos son crticos para la conformacin de las protenas y pptidos, y su estudio ofrece varios beneficios:

Mayores fluctuaciones pueden sugerir regiones ms dinmicas o flexibles, mientras que fluctuaciones menores pueden indicar estructuras ms rgidas o estables.

2. **Influencia de la temperatura**: Comparar los ngulos dihedros a diferentes temperaturas ayuda a comprender cmo el calor afecta la conformacin de las protenas. Por ejemplo, a temperaturas ms altas, las molculas tienen ms energa trmica, lo que puede aumentar la movilidad y cambiar los ngulos dihedros.

3. **Conformaciones preferidas y transiciones**: Los residuos de aminocidos en protenas tienen conformaciones preferidas que se representan comnmente en el mapa de Ramachandran. Estudiar estos ngulos a diferentes temperaturas puede revelar cambios en las conformaciones preferidas o mostrar cmo los residuos pasan de una conformacin a otra.

Interpretando las grficas especficamente:

- **ALA-2**: Se observa que los ngulos dihedros a 298K y 400K fluctan alrededor de diferentes valores medios, lo que sugiere que el cambio de temperatura puede haber alterado la conformacin promedio o la dinmica de las transiciones entre conformaciones. ```python

```python
def plot_ramachandran_from_dat(file_name, title):
    """Genera un diagrama de Ramachandran a partir de un archivo .dat."""
    data = np.loadtxt(file_name)  # Carga los datos del archivo
    phi = data[:, 1]  # Asume que phi est en la segunda columna
    psi = data[:, 2]  # Asume que psi est en la tercera columna
    
    plt.figure(figsize=(8, 6))
    plt.hexbin(phi, psi, gridsize=50, cmap='inferno', bins='log')
    plt.colorbar(label='Log10(N)')
    plt.xlabel('ngulo Phi ()', size = 20)
    plt.ylabel('ngulo Psi ()', size = 20)
    plt.title(title, size = 20)
    plt.xlim([-180, 180])
    plt.ylim([-180, 180])
    plt.axhline(0, color='grey', lw=0.5)
    plt.axvline(0, color='grey', lw=0.5)
    plt.show()

# Generar diagramas de Ramachandran para cada aminocido
plot_ramachandran_from_dat('tarea/5-run_optional/ala-2_298k.dat', 'Ramachandran Plot Alanine1')
plot_ramachandran_from_dat('tarea/5-run_optional/gln-3_298k.dat', 'Ramachandran Plot Glutamina')
plot_ramachandran_from_dat('tarea/5-run_optional/ala-4_298k.dat', 'Ramachandran Plot Alanine2')
```
