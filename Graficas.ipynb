```python
def plot_dihedral(data_298k, data_400k, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data_298k['Frame'], data_298k['Phi'], label='298K Phi', color='blue')
    plt.plot(data_298k['Frame'], data_298k['Psi'], label='298K Psi', color='skyblue')
    plt.plot(data_400k['Frame'], data_400k['Phi'], label='400K Phi', color='red')
    plt.plot(data_400k['Frame'], data_400k['Psi'], label='400K Psi', color='salmon')
    plt.title(f'Ángulos diedros de {title}', size = 20)
    plt.xlabel('fentosegundos', size = 20)
    plt.ylabel('Ángulo (grados)', size = 20)
    plt.legend()
    plt.grid(False)
    plt.show()

# Cargar los datos
ala_2_298k_data = pd.read_csv('tarea/4-analysis/ala-2_298k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])
ala_2_400k_data = pd.read_csv('tarea/4-analysis/ala-2_400k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])
gln_3_298k_data = pd.read_csv('tarea/4-analysis/gln-3_298k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])
gln_3_400k_data = pd.read_csv('tarea/4-analysis/gln-3_400k.dat', delim_whitespace=True, header=None, names=['Frame', 'Phi', 'Psi'])

# Graficar
plot_dihedral(ala_2_298k_data, ala_2_400k_data, 'ALA-2')
plot_dihedral(gln_3_298k_data, gln_3_400k_data, 'GLN-3')
```

Las gráficas proporcionadas muestran los ángulos dihedros φ (Phi) y ψ (Psi) para dos residuos diferentes (ALA-2 y GLN-3) de un tripéptido AQA a dos temperaturas distintas (298K y 400K). Estos ángulos son críticos para la conformación de las proteínas y péptidos, y su estudio ofrece varios beneficios:

1. **Flexibilidad y dinámica de la estructura**: Observar la variabilidad en los ángulos dihedros puede indicar la flexibilidad de ciertas regiones de la molécula. Mayores fluctuaciones pueden sugerir regiones más dinámicas o flexibles, mientras que fluctuaciones menores pueden indicar estructuras más rígidas o estables.

2. **Influencia de la temperatura**: Comparar los ángulos dihedros a diferentes temperaturas ayuda a comprender cómo el calor afecta la conformación de las proteínas. Por ejemplo, a temperaturas más altas, las moléculas tienen más energía térmica, lo que puede aumentar la movilidad y cambiar los ángulos dihedros.

3. **Conformaciones preferidas y transiciones**: Los residuos de aminoácidos en proteínas tienen conformaciones preferidas que se representan comúnmente en el mapa de Ramachandran. Estudiar estos ángulos a diferentes temperaturas puede revelar cambios en las conformaciones preferidas o mostrar cómo los residuos pasan de una conformación a otra.

Interpretando las gráficas específicamente:

- **ALA-2**: Se observa que los ángulos dihedros a 298K y 400K fluctúan alrededor de diferentes valores medios, lo que sugiere que el cambio de temperatura puede haber alterado la conformación promedio o la dinámica de las transiciones entre conformaciones. A 400K, las fluctuaciones parecen ser más amplias, lo que puede indicar mayor flexibilidad o una barrera de energía más baja para las transiciones conformacionales debido al incremento de energía térmica.

- **GLN-3**: Similar a ALA-2, GLN-3 muestra diferencias en las fluctuaciones de los ángulos dihedros entre 298K y 400K. Estas diferencias pueden tener implicaciones importantes para la estructura y dinámica del residuo de glutamina y, por extensión, para la conformación general del tripéptido.

En ambos casos, el estudio de los ángulos dihedros contribuye a una mejor comprensión de la estructura secundaria y la flexibilidad local del tripéptido bajo diferentes condiciones de temperatura, lo que es fundamental para el diseño de péptidos y proteínas con propiedades físicas y funciones biológicas específicas.

```python

def plot_ramachandran_from_dat(file_name, title):
    """Genera un diagrama de Ramachandran a partir de un archivo .dat."""
    data = np.loadtxt(file_name)  # Carga los datos del archivo
    phi = data[:, 1]  # Asume que phi está en la segunda columna
    psi = data[:, 2]  # Asume que psi está en la tercera columna
    
    plt.figure(figsize=(8, 6))
    plt.hexbin(phi, psi, gridsize=50, cmap='inferno', bins='log')
    plt.colorbar(label='Log10(N)')
    plt.xlabel('Ángulo Phi (φ)', size = 20)
    plt.ylabel('Ángulo Psi (ψ)', size = 20)
    plt.title(title, size = 20)
    plt.xlim([-180, 180])
    plt.ylim([-180, 180])
    plt.axhline(0, color='grey', lw=0.5)
    plt.axvline(0, color='grey', lw=0.5)
    plt.show()

# Generar diagramas de Ramachandran para cada aminoácido
plot_ramachandran_from_dat('tarea/5-run_optional/ala-2_298k.dat', 'Ramachandran Plot Alanine1')
plot_ramachandran_from_dat('tarea/5-run_optional/gln-3_298k.dat', 'Ramachandran Plot Glutamina')
plot_ramachandran_from_dat('tarea/5-run_optional/ala-4_298k.dat', 'Ramachandran Plot Alanine2')
```
